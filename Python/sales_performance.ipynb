```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("train.csv")
```

```python
data['Postal Code'].fillna(data['Postal Code'].mean(), inplace =True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
cat_count=data['Category'].value_counts()
cat_count
```

```python
sub_count=data['Sub-Category'].value_counts()
```

```python
sub_count
```

```python
plt.figure(figsize=(5,5))
```

```python
sub_count.plot(kind='bar',color='pink')
plt.xlabel('sub_category')
plt.ylabel('count')
plt.title('sub_categories count')
plt.show()
```

```python
sub_sales = data.groupby('Sub-Category')['Sales'].mean()
sub_sales
```

```python
plt.figure(figsize=(5, 5))
sub_sales.plot(kind='bar',color='green')
plt.xlabel('sub_category')
plt.ylabel('Average Sales')
plt.title('Average Sales by Sub-Category')
plt.show()
```

```python
Ca_sales = data.groupby('Category')['Sales'].mean()
plt.figure(figsize=(5, 5))
Ca_sales.plot(kind='bar',color='green')
plt.xlabel('Category')
plt.ylabel('Average Sales')
plt.title('Average Sales by Category')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d/%m/%Y')
```

```python
data['Year'] = data['Ship Date'].dt.year
data['Month'] = data['Ship Date'].dt.month
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=data, x='Ship Date', y='Sales', estimator='sum', ci=None)
plt.title('Sales Over Time')
plt.xlabel('Ship Date')
plt.ylabel('Sales')
plt.show()
```
